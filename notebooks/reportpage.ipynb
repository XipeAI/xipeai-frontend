```python
# Import necessary libraries
import os
import pydicom
from pydicom.dataset import Dataset, FileDataset
from datetime import datetime
import numpy as np
```

```python
# Function to generate a new DICOM file with tumor information
def create_dicom_report(output_path, patient_info, tumor_info):
    # Create a new DICOM file
    file_meta = Dataset()
    file_meta.MediaStorageSOPClassUID = pydicom.uid.generate_uid()
    file_meta.MediaStorageSOPInstanceUID = pydicom.uid.generate_uid()
    file_meta.ImplementationClassUID = pydicom.uid.generate_uid()
    
    ds = FileDataset(output_path, {}, file_meta=file_meta, preamble=b"\0" * 128)
    
    # Add patient information
    ds.PatientName = patient_info['name']
    ds.PatientID = patient_info['id']
    ds.PatientBirthDate = patient_info['birth_date']
    ds.StudyDate = datetime.now().strftime('%Y%m%d')
    ds.StudyTime = datetime.now().strftime('%H%M%S')
    ds.Modality = 'OT'  # Other
    ds.StudyDescription = 'Tumor Analysis Report'
    
    # Add custom tumor information
    ds.add_new(0x0008103E, 'LO', 'Tumor Information')
    
    for idx, tumor in enumerate(tumor_info, start=1):
        ds.add_new(0x00200010 + idx, 'LO', f"Tumor {idx}: Volume={tumor['volume']:.2f}mm³, Density={tumor['density']:.2f}")
    
    # Save the DICOM file
    ds.save_as(output_path)
    print(f"Report saved as {output_path}")
```

```python
# Example patient information
patient_info = {
    'name': 'John Doe',
    'id': '123456',
    'birth_date': '19800101'
}

# Example tumor information extracted from the analysis
tumor_info = [
    {'volume': 1234.56, 'density': 0.98},
    {'volume': 2345.67, 'density': 1.05},
    # Add more tumors as needed
]
```

```python
# Output path for the new DICOM report
output_path = '/Users/calin/Desktop/tumors/tumor_report.dcm'

# Create the DICOM report
create_dicom_report(output_path, patient_info, tumor_info)
```

```python
import pydicom

def display_dicom_metadata(dicom_path):
    # Load the DICOM file
    ds = pydicom.dcmread(dicom_path)
    
    # Display the metadata
    print("Patient Name:", ds.PatientName)
    print("Patient ID:", ds.PatientID)
    print("Patient Birth Date:", ds.PatientBirthDate)
    print("Study Date:", ds.StudyDate)
    print("Study Time:", ds.StudyTime)
    print("Modality:", ds.Modality)
    print("Study Description:", ds.StudyDescription)
    print("\nTumor Information:")
    
    # Extract and display custom tumor information
    for elem in ds:
        if elem.tag.group == 0x0020 and elem.tag.element >= 0x0010:
            print(elem.value)
    
# Path to the DICOM report
dicom_report_path = '/Users/calin/Desktop/tumors/tumor_report.dcm'

# Display the metadata
display_dicom_metadata(dicom_report_path)
```
